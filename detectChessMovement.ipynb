```python
def predict_on_large_image(model, image, image_size=(224, 224), overlap_ratio=0.25):
    large_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    original_h, original_w, _ = large_image.shape

    patch_h, patch_w = image_size
    stride_h = int(patch_h * (1 - overlap_ratio))
    stride_w = int(patch_w * (1 - overlap_ratio))

    full_prediction = np.zeros((original_h, original_w), dtype=np.float32)
    overlap_counter = np.zeros((original_h, original_w), dtype=np.float32)

    with torch.no_grad():
        for y in range(0, original_h, stride_h):
            for x in range(0, original_w, stride_w):
                y_end = min(y + patch_h, original_h)
                x_end = min(x + patch_w, original_w)
                y_start = y_end - patch_h
                x_start = x_end - patch_w

                if y_start < 0: y_start = 0
                if x_start < 0: x_start = 0

                patch = large_image[y_start:y_end, x_start:x_end]

                processed_patch = cv2.resize(patch, (image_size[1], image_size[0]))
                processed_patch = np.transpose(processed_patch, (2, 0, 1)).astype(np.float32)
                processed_patch = torch.Tensor(processed_patch) / 255.0
                processed_patch = processed_patch.unsqueeze(0)
                processed_patch = processed_patch.to('cuda')

                output_logits = model(processed_patch)
                output_mask = torch.sigmoid(output_logits).squeeze().cpu().numpy()

                output_mask_resized = cv2.resize(output_mask, (x_end - x_start, y_end - y_start))

                full_prediction[y_start:y_end, x_start:x_end] += output_mask_resized
                overlap_counter[y_start:y_end, x_start:x_end] += 1

    full_prediction /= overlap_counter
    final_mask = (full_prediction > 0.5).astype(np.uint8) * 255

    return large_image, final_mask

class conv_block(nn.Module):
  def __init__(self, in_c, out_c):
    super().__init__()
    self.conv1 = nn.Conv2d(in_c, out_c, kernel_size=3, padding=1)
    self.bn1 = nn.BatchNorm2d(out_c)
    self.conv2 = nn.Conv2d(out_c, out_c, kernel_size=3, padding=1)
    self.bn2 = nn.BatchNorm2d(out_c)
    self.relu = nn.ReLU()
  def forward(self, inputs):
    x = self.conv1(inputs)
    x = self.bn1(x)
    x = self.relu(x)
    x = self.conv2(x)
    x = self.bn2(x)
    x = self.relu(x)
    return x

class encoder_block(nn.Module):
  def __init__(self, in_c, out_c):
    super().__init__()
    self.conv = conv_block(in_c, out_c)
    self.pool = nn.MaxPool2d((2, 2))
  def forward(self, inputs):
    x = self.conv(inputs)
    p = self.pool(x)
    return x, p

class decoder_block(nn.Module):
  def __init__(self, in_c, out_c):
    super().__init__()
    self.up = nn.ConvTranspose2d(in_c, out_c, kernel_size=2, stride=2, padding=0)
    self.conv = conv_block(out_c+out_c, out_c)
  def forward(self, inputs, skip):
    x = self.up(inputs)
    x = torch.cat([x, skip], axis=1)
    x = self.conv(x)
    return x

class build_unet(nn.Module):
  def __init__(self):
    super().__init__()
    """ Encoder """
    self.e1 = encoder_block(3, 64)
    self.e2 = encoder_block(64, 128)
    self.e3 = encoder_block(128, 256)
    self.e4 = encoder_block(256, 512)
    """ Bottleneck """
    self.b = conv_block(512, 1024)
    """ Decoder """
    self.d1 = decoder_block(1024, 512)
    self.d2 = decoder_block(512, 256)
    self.d3 = decoder_block(256, 128)
    self.d4 = decoder_block(128, 64)
    """ Classifier """
    self.outputs = nn.Conv2d(64, 1, kernel_size=1, padding=0)
  def forward(self, inputs):
    """ Encoder """
    s1, p1 = self.e1(inputs)
    s2, p2 = self.e2(p1)
    s3, p3 = self.e3(p2)
    s4, p4 = self.e4(p3)
    """ Bottleneck """
    b = self.b(p4)
    """ Decoder """
    d1 = self.d1(b, s4)
    d2 = self.d2(d1, s3)
    d3 = self.d3(d2, s2)
    d4 = self.d4(d3, s1)
    """ Classifier """
    outputs = self.outputs(d4)
    return outputs

class SegmentationModel(nn.Module):
  def __init__(self):
    super(SegmentationModel,self).__init__()
    self.arc = build_unet()

  def forward(self, images, masks = None):
    logits = self.arc(images)

    if masks != None:
      pass

    return logits
```

```python
def canny(img, ksize=9):
  img_blur = cv2.GaussianBlur(img, (ksize, ksize), 0)
  median = np.median(img)
  image_canny = cv2.Canny(img_blur, median*0.8, median*1.2)
  return image_canny, int(median*1.2)
```

```python
def get_grid_limits(all_lines, threshold=40):
    """
    Input: all_lines = list of [x1, y1, x2, y2]
    Output: x_limits, y_limits (sorted lists of grid coordinates)
    """
    raw_x = []
    raw_y = []

    # 1. Separate X and Y coordinates based on line orientation
    for line in all_lines:
        x1, y1, x2, y2 = line[0]
        if abs(x2 - x1) > abs(y2 - y1):
            # Horizontal line: we care about Y
            raw_y.append((y1 + y2) // 2)
        else:
            # Vertical line: we care about X
            raw_x.append((x1 + x2) // 2)

    # 2. Function to merge nearby duplicate lines
    def merge_coordinates(coords):
        coords.sort()
        if not coords: return []

        merged = []
        current_group = [coords[0]]

        for val in coords[1:]:
            # If the next line is close to the current one (within threshold), group it
            if val - current_group[-1] < threshold:
                current_group.append(val)
            else:
                # Average the group to get one clean line
                merged.append(int(np.mean(current_group)))
                current_group = [val]

        merged.append(int(np.mean(current_group)))
        return merged

    # 3. Get the final lists
    x_limits = merge_coordinates(raw_x)
    y_limits = merge_coordinates(raw_y)

    return x_limits, y_limits

# --- Example Usage ---
# Assuming 'lines' is your list of raw [x1, y1, x2, y2] data
x_limits, y_limits = get_grid_limits(lines)

print("X Limits (Vertical Lines):", x_limits)
print("Y Limits (Horizontal Lines):", y_limits)
```

```python
print(len(frames))
board_states = [] # Initialize list to store all detected unique stable board states

# To keep track of the most recent *stable* board state for comparison and plotting.
last_processed_stable_board_state = None
last_processed_stable_results = None

# To count consecutive frames without detected movement.
consecutive_static_frames = 0
STABILITY_THRESHOLD = 5 # Number of static frames to confirm stability

# Flag to know if we are currently in a period of movement
in_movement_phase = False

# --- Step 1: Handle the initial board state ---
if len(frames) > 0:
    # Warp the initial frame to get the board perspective
    image_resized_initial = cv2.warpPerspective(frames[0], M, (width + 2 * threshold, height + 2 * threshold))
    results_initial = model(image_resized_initial)
    initial_board_state_dict = {}
    for idx, box in enumerate(results_initial[0].boxes):
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        cls = int(box.cls[0])
        conf = float(box.conf[0]) # Get confidence score
        class_name = results_initial[0].names[cls]
        if class_name not in initial_board_state_dict:
            initial_board_state_dict[class_name] = []
        initial_board_state_dict[class_name].append((int((x2+x1)//2-60), int((y2-20)-60), conf)) # Store confidence

    board_states.append(initial_board_state_dict)
    last_processed_stable_board_state = initial_board_state_dict
    last_processed_stable_results = results_initial

    # Display the initial board state
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(results_initial[0].plot(), cv2.COLOR_BGR2RGB))
    plt.title(f"Detected Objects in Initial Frame (Processed from frames[0])")
    plt.show()

# --- Step 2: Iterate through subsequent frames ---
for i in range(1, len(frames)):
    # Check for movement between the current frame and the immediately preceding one
    movement_detected_in_pair = isDifferent(frames[i-1], frames[i], 140)

    if movement_detected_in_pair:
        # If movement is detected, we enter (or continue) the movement phase.
        in_movement_phase = True
        consecutive_static_frames = 0 # Reset static counter as movement is happening
    else:
        # No movement detected in this frame pair.
        consecutive_static_frames += 1

        if in_movement_phase and consecutive_static_frames >= STABILITY_THRESHOLD:
            # We were in a movement phase, and now the board has been static for long enough.
            # This signifies the *end* of a move and the establishment of a *new stable state*.

            # Process the current frame (frames[i]) to get the new stable board state.
            image_resized = cv2.warpPerspective(frames[i], M, (width + 2 * threshold, height + 2 * threshold))
            results = model(image_resized)
            current_board_state_dict = {}
            for idx, box in enumerate(results[0].boxes):
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                cls = int(box.cls[0])
                conf = float(box.conf[0]) # Get confidence score
                class_name = results[0].names[cls]
                if class_name not in current_board_state_dict:
                    current_board_state_dict[class_name] = []
                current_board_state_dict[class_name].append((int((x2+x1)//2)-60, int((y2-20))-60, conf)) # Store confidence

            # If this new stable state is different from the last one we recorded, add it.
            if current_board_state_dict != last_processed_stable_board_state:
                board_states.append(current_board_state_dict)
                last_processed_stable_board_state = current_board_state_dict
                last_processed_stable_results = results

                # Display the detected objects in the new stable state
                plt.figure(figsize=(10,10))
                plt.imshow(cv2.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB))
                plt.title(f"Detected Objects in New Stable State (Processed from frames[{i}])")
                plt.show()

            in_movement_phase = False # Exit movement phase, now in a stable phase
            consecutive_static_frames = 0 # Reset for the next potential movement detection

print(f"Collected {len(board_states)} unique board states.")
```

### 3. Temporal Consistency (Move Legality - Addressing PGN Warnings)

The warnings from `generate_pgn` ("No legal move found") are a key indicator of issues. This means that the transition from one `board_matrix` to the next, as interpreted by `python-chess`, does not correspond to a single, legal chess move.

**Possible reasons for PGN warnings:**

*   **Detection Errors**: Pieces are misidentified, mislocated, or missed entirely in one or both frames of a transition.
*   **Temporal Resolution**: The frames you've selected as 'stable states' might still capture an intermediate or ambiguous state rather than a clear before-and-after of a single move.
*   **Ambiguous Moves**: Some real-world moves might be hard for a simple difference analysis to map to a unique `python-chess` move without additional context (e.g., castling, en passant, promotions).
*   **Initial FEN Assumptions**: The `generate_pgn` function makes assumptions about the castling rights and turn to move if the game doesn't start from a standard position. Incorrect assumptions can lead to false negatives for legal moves.

```python
import chess
import chess.pgn

# --- Piece Definitions ---
# Mapping integer IDs to python-chess constants
# 1-6: Black (P, B, N, R, Q, K)
# 11-16: White (P, B, N, R, Q, K)
ID_TO_PIECE = {
    1: chess.Piece(chess.PAWN, chess.BLACK),
    2: chess.Piece(chess.BISHOP, chess.BLACK),
    3: chess.Piece(chess.KNIGHT, chess.BLACK),
    4: chess.Piece(chess.ROOK, chess.BLACK),
    5: chess.Piece(chess.QUEEN, chess.BLACK),
    6: chess.Piece(chess.KING, chess.BLACK),
    11: chess.Piece(chess.PAWN, chess.WHITE),
    12: chess.Piece(chess.BISHOP, chess.WHITE),
    13: chess.Piece(chess.KNIGHT, chess.WHITE),
    14: chess.Piece(chess.ROOK, chess.WHITE),
    15: chess.Piece(chess.QUEEN, chess.WHITE),
    16: chess.Piece(chess.KING, chess.WHITE),
    0: None
}

def matrix_to_fen(matrix):
    """
    Converts an 8x8 integer matrix into a FEN string (piece placement only).
    Row 0 is Rank 8, Row 7 is Rank 1.
    """
    rows = []
    for r in range(8):
        empty_count = 0
        row_str = ""
        for c in range(8):

            piece_id = matrix[r][c]
            piece = ID_TO_PIECE.get(piece_id)

            if piece is None:
                empty_count += 1
            else:
                if empty_count > 0:
                    row_str += str(empty_count)
                    empty_count = 0
                row_str += piece.symbol()

        if empty_count > 0:
            row_str += str(empty_count)
        rows.append(row_str)

    # print(rows)

    return "/".join(rows)

def match_state(chess_board, target_matrix):
    """
    Checks if the current chess.Board matches the target 8x8 integer matrix.
    """
    current_fen_placement = chess_board.fen().split(' ')[0]
    target_fen_placement = matrix_to_fen(target_matrix)
    return current_fen_placement == target_fen_placement

def try_detect_move(board, next_matrix):
    """
    Helper to try finding a move on the current board configuration.
    Returns the move if found, else None.
    """
    for move in board.legal_moves:
        board.push(move)
        is_match = match_state(board, next_matrix)
        board.pop()
        if is_match:
            return move
    return None

def generate_pgn(board_states):
    """
    Converts a list of 8x8 integer matrices into a PGN string.
    """
    if not board_states:
        return ""

    game = chess.pgn.Game()

    # 1. Determine Start Position and Turn Color
    start_fen = matrix_to_fen(board_states[0])
    STANDARD_START = "rnbqkbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBQKBNR"

    if start_fen == STANDARD_START:
        board = chess.Board() # Standard start
    else:
        # Mid-game start logic
        # FIX: We use "KQkq" instead of "-" to assume castling rights are available if pieces are in place.
        # python-chess automatically invalidates rights if the Rooks/Kings aren't on start squares.
        board = chess.Board(start_fen + " w KQkq - 0 1") # Assume White first

        if len(board_states) > 1:
            move_white = try_detect_move(board, board_states[1])

            if move_white:
                game.setup(board)
            else:
                # Try Black
                # FIX: Also use "KQkq" for Black detection
                board = chess.Board(start_fen + " b KQkq - 0 1")
                move_black = try_detect_move(board, board_states[1])

                if move_black:
                    game.setup(board)
                else:
                    # Default to White if ambiguous, maintaining permissions
                    board = chess.Board(start_fen + " w KQkq - 0 1")
                    game.setup(board)
        else:
             game.setup(board)

    node = game

    # 2. Iterate through states
    for i in range(len(board_states) - 1):
        current_matrix = board_states[i]
        next_matrix = board_states[i+1]

        # Skip duplicate frames
        '''
        if current_matrix == next_matrix:
            continue
        '''

        found_move = try_detect_move(board, next_matrix)

        if found_move:
            node = node.add_variation(found_move)
            board.push(found_move)
        else:
            print(f"Warning: No legal move found between state {i} and {i+1}")

    # Export PGN without headers
    exporter = chess.pgn.StringExporter(headers=False, variations=True, comments=True)
    pgn_output = game.accept(exporter)

    return pgn_output.replace(" *", "").strip()
```
